```python
import json
import os
import zipfile
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from nltk.corpus import stopwords
import pandas as pd
#import panda as pd
```

```python
papers = [] 
directory = r'./data'
for entry in os.scandir(directory):
    if (entry.path.endswith('.json')):
        papers.append(entry)
```

```python
titles = np.array([])         #training data
performance = np.array([])    #labels
for i in papers:
    f = open(i,'r',encoding='cp850')
    k = f.read()
    y = json.loads(k)
    l= y["title"]
    m = y["thread"]["performance_score"]
#     if int(l) > 5:
    performance  = np.append(performance,m)
    titles = np.append(titles,l)
```

```python
#tokenization
training_size = int(len(papers)/2) 

tokenizer = Tokenizer(num_words=14000, oov_token="<00V>")
tokenizer.fit_on_texts(titles[0:training_size] )
word_index = tokenizer.word_index
```

```python
#splitting data for testing vs. training
training_size = int(len(papers)/2) 

training_sentences = tokenizer.texts_to_sequences(titles[0:training_size])
testing_sentences = tokenizer.texts_to_sequences(titles[training_size:])

padded_training = pad_sequences(training_sentences, padding='post')
padded_testing = pad_sequences(testing_sentences, padding='post')
training_labels = performance[0:training_size]
testing_labels = performance[training_size:]
```

```python
vocab_size = 10000

model = keras.Sequential([
    keras.layers.Embedding(vocab_size, 16),
    keras.layers.GlobalAveragePooling1D(),
    keras.layers.Dense(24,activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')
])

model.compile(optimizer='adam', loss="sparse_categorical_crossentropy", metrics=['accuracy'])

model.fit(padded_training, training_labels, epochs=30, validation_data=(padded_testing, testing_labels), verbose=2)
```
